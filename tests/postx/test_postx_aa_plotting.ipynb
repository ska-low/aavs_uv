```python
"""Test aa_plotter plotting routines."""
import pylab as plt
from aa_uv.io import hdf5_to_uvx
from aa_uv.postx import ApertureArray
from aa_uv.utils import from aa_uv.utils import get_test_data


def test_aa_plotter():
    """Test plotting routines."""
    uvx = hdf5_to_uvx(get_test_data('aavs2_1x1000ms/correlation_burst_204_20230823_21356_0.hdf5'), telescope_name='aavs2')
    aa = ApertureArray(uvx)

    plt.figure("ANT")
    aa.plotting.plot_antennas()

    plt.figure("CORR")
    aa.plotting.plot_corr_matrix()

    #plt.figure("CORR 4x")
    aa.plotting.plot_corr_matrix_4pol()

    plt.figure("UV DIST")
    aa.plotting.plot_uvdist_amp()

test_aa_plotter()
```
